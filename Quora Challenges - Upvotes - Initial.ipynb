```python
# This is a solution for the Quora challenge named "Upvotes" at https://www.quora.com/challenges
import numpy as np
N = [int(x) for x in input("Enter the amount of upvotes per day: ").split()]
```

```python
K = int(input("Enter the size of the window to be analyzed: "))

def upvote_window_check(x,y):
    if len(x) > 100000 or len(x) < 1:
        print ("Data array is not valid, please enter a data array with at least 1 day and at most 100000 days")
    else:
        print ("Days = ", len(x))
    if y > len(x) or y < 1:
        print ("Window is invalid, please enter a valid window size between one and the amount of days")
    else:
        print ("Window size = ",y)
        
def shift_window(window,initial):
    window = np.append(window, N[initial+K-1])
    window = np.delete(window,0,0)
    return window

def start_days_analysis(N,K):
    initial = 0
    i = 0
    upvote_window_check(N, K)
    window = np.array(N[initial:initial+K])
    single_window_analysis(initial, i, window)

def single_window_analysis(initial, i, window):
    final = i + K - 1
    nonD = 0
    nonI = 0
    nonD_net = 0
    if initial + K > len(N):
        print ("Analysis Complete")
    else:
        window_chain_start(initial, final, i, window, nonD, nonI)

def window_chain_start(initial, final, i, window, nonD, nonI):
    j = 1
    if i == final:
        calculate_net(initial, window, nonD, nonI)
    else:
        unknown_chain(initial, final, i, j, window, nonD, nonI)

def nonD_chain(initial, final, i, j, window, nonD, nonI):
    o = j-1
    if i+j > final:
        i+=1
        window_chain_start(initial, final, i, window, nonD, nonI)
    elif window[i+o] <= window[i+j]:
        nonD+=1
        j+=1
        nonD_chain(initial, final, i, j, window, nonD, nonI)
    else:
        i+=1
        window_chain_start(initial, final, i, window, nonD, nonI)

def nonI_chain(initial,final, i, j, window, nonD, nonI):
    o = j-1
    if i+j > final:
        i+=1
        window_chain_start(initial, final, i, window, nonD, nonI)
    elif window[i+o] >= window[i+j]:
        nonI+=1
        j+=1
        nonI_chain(initial, final, i, j, window, nonD, nonI)
    else:
        i+=1
        window_chain_start(initial, final, i, window, nonD, nonI)

def unknown_chain(initial, final, i, j, window, nonD, nonI):
    if i+j > final:
        i+=1
        window_chain_start(initial, final, i, window, nonD, nonI)
    else:
        if window[i] < window[i+j]:
            nonD+=1
            j+=1
            nonD_chain(initial, final, i, j, window, nonD, nonI)
        elif window[i] > window[i+j]:
            nonI+=1
            j+=1
            nonI_chain(initial, final, i, j, window, nonD, nonI)
        else:
            j+=1
            unknown_chain(initial, final, i, j, window, nonD, nonI)

def calculate_net(initial, window, nonD, nonI):
    nonD_net = nonD - nonI
    print (nonD_net)
    initial +=1
    i = 0
    if initial == len(N) - K + 1:
        print ("Analysis Complete")
    else:
        window = shift_window(window, initial)
        single_window_analysis(initial, i, window)
start_days_analysis(N,K)
```
